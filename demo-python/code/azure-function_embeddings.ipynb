```python
import requests
import json
```

```python

url = "https://maggetimgembd001.azurewebsites.net/api/GetImageEmbeddings"

payload = json.dumps({
  "values": [
    {
      "data": {
        "imageUrl": "https://cdn.ecommercedns.uk/files/9/212109/9/5235269/classic-top-hat-red.jpg"
      },
      "recordId": "1"
    },
    {
      "data": {
        "imageUrl": "https://assets.kogan.com/images/blossomcostumes/BSM-AC1025wSMF_48261/1-b2b1174f6d-red-sequin-top-hat-48261.jpg"
      },
      "recordId": "2"
    },
    {
    "data": {
        "imageUrl": "https://godzillastorage.blob.core.windows.net/searchvectordemo/Picture999.jpeg"
    }, 
    "recordId": "4"
  }
  ]
})
headers = {
  'Content-Type': 'application/json'
}
```

```python

response = requests.request("POST", url, headers=headers, data=payload)

print(response.text)
```

```python
# read "values" from response
json_response = json.loads(response.text)
values = json_response['values']
```

```python
type(values)

for value in values: 
    print(list(value.keys()))
    print("recordId: " + value['recordId'])
    print(f"data\\vector: \n\tlen={len(value['data']['vector'])}, \n\tvalues: {value['data']['vector'][:10]} ...)") 
    print('########\n')
```
